**1- Utilize a função abaixo caso a imagem venha no formato RGB**

```python
from google.colab.patches import cv2_imshow
import cv2

#essa função mostra a imagem no google colab
def mostrar(img):
  fig = plt.gcf()
  fig.set_size_inches(16,10)
  plt.axis('off') #nao mostrar valor no eixo x e y
  cv2_imshow(img)

resultado = cv2.imread('/content/runs/detect/predict/NOME DO ARQUIVO') 
mostrar(resultado)
```

**2- Utilize a função abaixo caso a imagem venha no formato BGR**

```python
def mostrar(img):
  fig = plt.gcf()
  fig.set_size_inches(16,10)
  plt.axis('off') #nao mostrar valor no eixo x e y
  plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
  plt.show()
```

```python
resultado = cv2.imread('/content/runs/detect/predict/NOME DO ARQUIVO')

mostrar(resultado)
```
